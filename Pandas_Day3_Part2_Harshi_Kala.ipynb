##Exercise 12: Duplicate Values
####Create duplicate records.
####Perform
####Find duplicates.
####Remove duplicates.
####Count duplicate rows.
####Keep only unique rows.

```python
missing = pd.DataFrame({
    "Name":["John","Alice","David","Sophia"],
    "Age":[25,np.nan,30,26],
    "Salary":[45000,52000,np.nan,47000]
})

print(missing)

print("\nMissing Values")
print(missing.isnull())

print("\nCount")
print(missing.isnull().sum())

print("\nDrop Missing")
print(missing.dropna())

print("\nFill with 0")
print(missing.fillna(0))

print("\nFill with Mean")
print(missing.fillna(missing.mean(numeric_only=True)))

print("\nForward Fill")
print(missing.ffill())

print("\nBackward Fill")
print(missing.bfill())
```

##Exercise 16: GroupBy
####Using
#####Department	Salary
#####IT	55000
#####IT	60000
#####HR	45000
#####HR	48000
#####Finance	70000
#####Finance	75000
####Find
#####Average salary
#####Maximum salary
#####Minimum salary
#####Employee count
#####Total salary

```python
dept = pd.Series([
    "IT","HR","IT","Finance","IT","HR","Sales"
])

print(dept.value_counts())

print("Most Common")
print(dept.value_counts().idxmax())

print("Least Common")
print(dept.value_counts().idxmin())
```

##Exercise 16: GroupBy
####Using
#####Department	Salary
#####IT	55000
#####IT	60000
#####HR	45000
#####HR	48000
#####Finance	70000
#####Finance	75000
####Find
#####Average salary
#####Maximum salary
#####Minimum salary
#####Employee count
#####Total salary

```python
group = pd.DataFrame({
    "Department":["IT","IT","HR","HR","Finance","Finance"],
    "Salary":[55000,60000,45000,48000,70000,75000]
})

print(group.groupby("Department")["Salary"].mean())

print(group.groupby("Department")["Salary"].max())

print(group.groupby("Department")["Salary"].min())

print(group.groupby("Department")["Salary"].count())

print(group.groupby("Department")["Salary"].sum())
```
