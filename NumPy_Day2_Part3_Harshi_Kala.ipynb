##Exercise 14: Hospital Patient Records
####Access records, slice data, reshape and join another day's records.

```python
import pandas as pd
import numpy as np
```

```python
# Load and access records
patients_day1 = pd.read_csv("hospital_patients_day1.csv")
print(patients_day1.info())

# Access records for a specific department
cardiology_patients = patients_day1[patients_day1["department"] == "Cardiology"]
print(cardiology_patients)
```

```python
# Slicing
# Row slicing by position
first_five = patients_day1.iloc[0:5]

# Column slicing
basic_info = patients_day1.loc[:, ["patient_id", "name", "age"]]

# Row + column slicing combined
young_patients_names = patients_day1.loc[patients_day1["age"] < 18, ["patient_id", "name", "age"]]
print(young_patients_names)
```

```python
# Reshape - pivot to count patients per department per status
dept_status_counts = patients_day1.pivot_table(
    index="department",
    columns="status",
    values="patient_id",
    aggfunc="count",
    fill_value=0
)
print(dept_status_counts)
```

```python
# Join with day 2 records
patients_day2 = pd.read_csv("hospital_patients_day2.csv")

# Concatenate both days into one combined record set (stacked rows)
all_records = pd.concat([patients_day1, patients_day2], ignore_index=True)

# Merge to see patients present on BOTH days (still admitted)
still_admitted = pd.merge(
    patients_day1, patients_day2,
    on="patient_id", suffixes=("_day1", "_day2"), how="inner"
)
print("Patients present on both days:\n",
      still_admitted[["patient_id", "name_day1", "status_day1", "status_day2"]])
```
