### Exercise 1: Importing Matplotlib

```python
import matplotlib.pyplot as plt
import matplotlib

print(matplotlib.__version__)

plt.figure()
plt.show()
```

### Exercise 2: Creating Your First Plot

```python
x = [1,2,3,4,5]
y = [10,20,30,40,50]

plt.plot(x, y)
plt.show()

plt.plot(x, y, color='red', linewidth=3, linestyle='--')
plt.show()
```

### Exercise 3: Line Plot

```python
months = [1,2,3,4,5,6]
sales = [1200,1500,1700,2100,2500,3000]

plt.plot(months, sales, marker='o', linestyle=':', markersize=10, markerfacecolor='green')
plt.show()
```

### Exercise 4: Titles and Labels

```python
plt.plot(months, sales, marker='o')
plt.title('Monthly Sales', fontsize=18, color='blue')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

### Exercise 5: Grid

```python
plt.plot(months, sales)
plt.grid(True, color='gray', linestyle='--', linewidth=0.7)
plt.show()
```

### Exercise 6: Multiple Lines

```python
months = [1,2,3,4,5,6]
sales_2024 = [1200,1500,1700,2100,2500,3000]
sales_2025 = [1400,1600,2000,2300,2700,3200]

plt.plot(months, sales_2024, color='blue', marker='o', label='2024')
plt.plot(months, sales_2025, color='orange', marker='s', label='2025')
plt.legend()
plt.grid(True)
plt.show()
```

### Exercise 7: Figure Size

```python
for size in [(6,4), (10,5), (12,8)]:
    plt.figure(figsize=size)
    plt.plot(months, sales)
    plt.title(f'Figure size {size}')
    plt.show()
```

### Exercise 8: Bar Chart

```python
departments = ['HR','IT','Finance','Sales','Marketing']
employees = [45,90,35,70,60]

plt.bar(departments, employees, color=['red','blue','green','orange','purple'])
plt.title('Employees per Department')
plt.xlabel('Department')
plt.ylabel('Employees')
plt.show()
```

### Exercise 9: Horizontal Bar Chart

```python
plt.barh(departments, employees, color=['red','blue','green','orange','purple'])
plt.grid(True)
plt.show()
```

### Exercise 10: Pie Chart

```python
labels = ['Python','Java','C++','JavaScript','Go']
students = [45,25,15,5,10]

plt.pie(students, labels=labels, autopct='%1.1f%%', explode=[0.1,0,0,0,0], shadow=True, startangle=90)
plt.show()
```

### Exercise 11: Histogram

```python
import numpy as np
data = np.random.randn(100)

plt.hist(data, bins=10, color='skyblue', edgecolor='black')
plt.show()

plt.hist(data, bins=20, color='salmon', edgecolor='black')
plt.show()
```

### Exercise 12: Scatter Plot

```python
study_hours = [1,2,3,4,5,6,7,8]
exam_marks = [35,40,50,55,65,70,85,90]

plt.scatter(study_hours, exam_marks, color='purple', s=100)
plt.title('Study Hours vs Exam Marks')
plt.grid(True)
plt.show()
```

### Exercise 13: Subplots

```python
fig, axs = plt.subplots(2, 2, figsize=(10,8))

axs[0,0].plot(months, sales)
axs[0,0].set_title('Line')

axs[0,1].bar(departments, employees)
axs[0,1].set_title('Bar')

axs[1,0].pie(students, labels=labels, autopct='%1.1f%%')
axs[1,0].set_title('Pie')

axs[1,1].scatter(study_hours, exam_marks)
axs[1,1].set_title('Scatter')

plt.tight_layout()
plt.show()
```

### Exercise 14: Legends

### Exercise 18: Employee Salary Analysis

```python
employee = ['A','B','C','D','E']
salary = [30000,45000,50000,60000,75000]

plt.plot(employee, salary, marker='o')
plt.title('Salary Line')
plt.show()

plt.bar(employee, salary, color='teal')
plt.title('Salary Bar')
plt.show()

plt.pie(salary, labels=employee, autopct='%1.1f%%')
plt.title('Salary Pie')
plt.show()

plt.hist(salary, bins=5, color='orange', edgecolor='black')
plt.title('Salary Histogram')
plt.show()

plt.scatter(employee, salary, color='red', s=100)
plt.title('Salary Scatter')
plt.show()
```

### Exercise 19: Student Marks

```python
students = ['S1','S2','S3','S4']
math = [80,90,70,85]
science = [75,85,95,80]
english = [90,70,85,75]

x = np.arange(len(students))
width = 0.25

plt.bar(x - width, math, width, label='Math')
plt.bar(x, science, width, label='Science')
plt.bar(x + width, english, width, label='English')
plt.xticks(x, students)
plt.legend()
plt.title('Grouped Bar Chart')
plt.show()

average = [(m+s+e)/3 for m,s,e in zip(math, science, english)]
plt.bar(students, average, color='green')
plt.title('Average Marks')
plt.show()

plt.pie(average, labels=students, autopct='%1.1f%%')
plt.title('Average Marks Pie')
plt.show()

plt.hist(math + science + english, bins=10, color='blue', edgecolor='black')
plt.title('Marks Histogram')
plt.show()
```

### Exercise 20: Weather Analysis

```python
days = [1,2,3,4,5,6,7]
temperature = [30,32,31,29,28,33,34]
humidity = [60,65,70,55,50,68,72]
rainfall = [0,5,0,10,20,0,0]

plt.plot(days, temperature, color='red', marker='o')
plt.title('Temperature')
plt.show()

plt.plot(days, humidity, color='blue', marker='o')
plt.title('Humidity')
plt.show()

plt.bar(days, rainfall, color='skyblue')
plt.title('Rainfall')
plt.show()

plt.scatter(temperature, humidity, color='purple')
plt.title('Temperature vs Humidity')
plt.show()

plt.hist(temperature, bins=5, color='orange', edgecolor='black')
plt.title('Temperature Histogram')
plt.show()
```

### Exercise 21: Sales Dashboard

```python
quarter = ['Q1','Q2','Q3','Q4']
product_a = [200,250,300,350]
product_b = [180,220,260,300]

fig, axs = plt.subplots(2, 2, figsize=(10,8))

axs[0,0].plot(quarter, product_a, marker='o', label='A')
axs[0,0].plot(quarter, product_b, marker='s', label='B')
axs[0,0].legend()
axs[0,0].set_title('Multiple Line Chart')

axs[0,1].bar(quarter, product_a, color='blue')
axs[0,1].set_title('Bar Chart')

axs[1,0].pie(product_a, labels=quarter, autopct='%1.1f%%')
axs[1,0].set_title('Pie Chart')

axs[1,1].scatter(product_a, product_b)
axs[1,1].set_title('Scatter')

plt.tight_layout()
plt.savefig('dashboard.png')
plt.show()
```

### Exercise 22: Banking Transactions

```python
months = [1,2,3,4,5,6]
deposits = [5000,6000,5500,7000,7500,8000]
withdrawals = [3000,3500,4000,3800,4200,4500]

plt.plot(months, deposits)
plt.title('Deposits')
plt.show()

plt.plot(months, deposits, label='Deposits', marker='o')
plt.plot(months, withdrawals, label='Withdrawals', marker='s')
plt.legend()
plt.title('Deposits vs Withdrawals')
plt.show()

plt.bar(months, deposits, color='green')
plt.title('Deposits Bar')
plt.show()

plt.pie([sum(deposits), sum(withdrawals)], labels=['Deposits','Withdrawals'], autopct='%1.1f%%')
plt.show()
```

### Exercise 27: Olympic Medal Analysis

```python
countries = ['USA','China','UK','Japan','Australia']
gold = [39,38,22,27,17]
silver = [41,32,21,14,7]
bronze = [33,18,22,17,22]

plt.bar(countries, gold, label='Gold', color='gold')
plt.bar(countries, silver, bottom=gold, label='Silver', color='silver')
plt.bar(countries, bronze, bottom=np.array(gold)+np.array(silver), label='Bronze', color='brown')
plt.legend()
plt.title('Stacked Medal Chart')
plt.show()

plt.pie(gold, labels=countries, autopct='%1.1f%%')
plt.title('Gold Medal Share')
plt.show()

plt.barh(countries, gold, color='gold')
plt.title('Gold Medals Horizontal')
plt.show()
```

### Exercise 28: Hospital Dashboard

```python
days = [1,2,3,4,5,6,7]
admissions = [20,25,22,30,28,26,24]
discharges = [18,20,21,25,26,22,23]
icu = [5,6,4,7,8,6,5]
bed_occupancy = [80,82,78,85,88,84,81]

fig, axs = plt.subplots(2, 2, figsize=(10,8))

axs[0,0].plot(days, admissions, marker='o')
axs[0,0].set_title('Admissions')

axs[0,1].plot(days, discharges, marker='o', color='green')
axs[0,1].set_title('Discharges')

axs[1,0].bar(days, icu, color='red')
axs[1,0].set_title('ICU Patients')

axs[1,1].plot(days, bed_occupancy, marker='o', color='purple')
axs[1,1].set_title('Bed Occupancy')

plt.tight_layout()
plt.show()
```
