##Exercise 15: Weather Dataset
####Perform indexing, slicing, reshape and joining operations on weather data.

```python
import pandas as pd
import numpy as np
```

```python
# Load and inspect
weather_w1 = pd.read_csv("weather_week1.csv")
print(weather_w1.head())
```

```python
# Indexing & slicing
# All readings for one city
jaipur_data = weather_w1[weather_w1["city"] == "Jaipur"]

# All readings for one date
july_3_data = weather_w1[weather_w1["date"] == "2026-07-03"]

# Using numpy for fast numeric indexing - convert temperature column to array
temps = weather_w1["temperature_c"].to_numpy()
hot_days_mask = temps > 35
print("Days above 35C:", temps[hot_days_mask])
```

```python
# Reshape - pivot so dates are rows, cities are columns, values = temperature
temp_pivot = weather_w1.pivot_table(
    index="date", columns="city", values="temperature_c"
)
print(temp_pivot)

# numpy reshape example: reshape flattened temp array into (days, cities) grid
n_days = weather_w1["date"].nunique()
n_cities = weather_w1["city"].nunique()
temp_grid = weather_w1["temperature_c"].to_numpy().reshape(n_days, n_cities)
print("Reshaped grid:\n", temp_grid)
```

```python
# Join week 1 and week 2 into a continuous dataset
weather_w2 = pd.read_csv("weather_week2.csv")

full_weather = pd.concat([weather_w1, weather_w2], ignore_index=True)
full_weather["date"] = pd.to_datetime(full_weather["date"])
full_weather = full_weather.sort_values(["city", "date"]).reset_index(drop=True)
print(full_weather.head(10))
```

```python
# Example: average temperature per city across both weeks
avg_temp_by_city = full_weather.groupby("city")["temperature_c"].mean().sort_values(ascending=False)
print(avg_temp_by_city)
```
