##Exercise 13: Product Inventory
####Access product names and quantities.
####Create copy, view and join another inventory dataset.

```python
import pandas as pd
import numpy as np
```

```python
inv_a = pd.read_csv("product_inventory_warehouseA.csv")
print(inv_a.head())
```

```python
# Access just names and quantities
names_qty = inv_a[["product_name", "quantity"]]
print(names_qty)
```

```python
# Access a single product's quantity by product_id
qty_p003 = inv_a.loc[inv_a['product_id'] == 'P003', 'quantity'].values[0]
print("Quantity of P003:", qty_p003)
```

```python
# .copy() creates an independent copy - edits don't affect the original

inv_a_copy = inv_a.copy()
inv_a_copy.loc[0, "quantity"] = 9999
print("Original unaffected:", inv_a.loc[0, "quantity"])
print("Copy changed:", inv_a_copy.loc[0, "quantity"])
```

```python
# A plain filtered slice can trigger SettingWithCopyWarning if you try to edit it;
# always use .copy() when you intend to modify a subset independently.
view_slice = inv_a[inv_a["category"] == "Electronics"]        # view-like, don't edit
safe_copy = inv_a[inv_a["category"] == "Electronics"].copy()   # safe to edit
```

```python
#Join with the second inventory dataset (Warehouse B)
inv_b = pd.read_csv("product_inventory_warehouseB.csv")
```

```python
# Inner join: only products present in BOTH warehouses
inner_join = pd.merge(inv_a, inv_b, on="product_id", suffixes=("_A", "_B"), how="inner")
```

```python
# Outer join: all products from both warehouses, missing values as NaN
outer_join = pd.merge(inv_a, inv_b, on="product_id", suffixes=("_A", "_B"), how="outer")
```

```python
# Combined total quantity across warehouses (only for common products)
inner_join["total_quantity"] = inner_join["quantity_A"] + inner_join["quantity_B"]
print(inner_join[["product_id", "product_name_A", "quantity_A", "quantity_B", "total_quantity"]])
```
