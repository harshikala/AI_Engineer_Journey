##9.⁠ ⁠Data Types
####Create arrays using i, b, u, f, c, m, M and U data types.
####Convert an integer array to float, boolean and string.
####Identify the datatype of a Unicode array.

```python
int_arr = np.array([1, 2, 3], dtype='i')
bool_arr = np.array([1, 0, 1], dtype='b')
unsigned_arr = np.array([10, 20, 30], dtype='u2')
float_arr = np.array([1.5, 2.5, 3.5], dtype='f')
complex_arr = np.array([1+2j, 3+4j], dtype='c')
timedelta_arr = np.array([5, 10, 15], dtype='m8[D]')
datetime_arr = np.array(['2025-01-01', '2025-12-31'], dtype='M8[D]')
unicode_arr = np.array(["Apple", "Banana", "Cherry"], dtype='U')

print("Integer:", int_arr)
print("Boolean:", bool_arr)
print("Unsigned:", unsigned_arr)
print("Float:", float_arr)
print("Complex:", complex_arr)
print("Timedelta:", timedelta_arr)
print("Datetime:", datetime_arr)
print("Unicode:", unicode_arr)
```

##10.⁠ ⁠Copy and View
####Create a copy of an array and modify the original.
####Create a view of an array and modify the original.
####Use .base to determine ownership of data.

```python
arr = np.array([10, 20, 30, 40])

# Copy
copy_arr = arr.copy()

arr[0] = 100

print("Original:", arr)
print("Copy:", copy_arr)

# View
view_arr = arr.view()

arr[1] = 200

print("\nOriginal:", arr)
print("View:", view_arr)

# Ownership using .base
print("\nCopy Base:", copy_arr.base)
print("View Base:", view_arr.base)
```

##13.⁠ ⁠Joining Arrays
####Join arrays using concatenate().
####Join 2D arrays row-wise and column-wise.
####Use stack(), vstack(), hstack() and dstack() and observe the output shapes.


```python
a = np.array([1,2,3])
b = np.array([4,5,6])

# concatenate
print("Concatenate:")
print(np.concatenate((a,b)))
```

```python
arr1 = np.array([[1,2],[3,4]])
arr2 = np.array([[5,6],[7,8]])

print("\nRow-wise Concatenate")
print(np.concatenate((arr1, arr2), axis=0))

print("\nColumn-wise Concatenate")
print(np.concatenate((arr1, arr2), axis=1))
```

```python
print("\nStack")
stack = np.stack((arr1, arr2))
print(stack)
print("Shape:", stack.shape)

print("\nVertical Stack")
v = np.vstack((arr1, arr2))
print(v)
print("Shape:", v.shape)

print("\nHorizontal Stack")
h = np.hstack((arr1, arr2))
print(h)
print("Shape:", h.shape)

print("\nDepth Stack")
d = np.dstack((arr1, arr2))
print(d)
print("Shape:", d.shape)
```

##Additional questions
####Create a 3x3 sales array.
####Check dimensions and shape.
####Access rows, columns and middle element.
####Slice rows.
####Create copy and view.
####Reshape to 1D.
####Join vertically and horizontally.
####Print datatype.

```python
sales = np.array([
    [101, 204, 303],
    [400, 503, 630],
    [720, 840, 990]
])

print("Sales Array:")
print(sales)

# Dimension and Shape

print("\nDimensions:", sales.ndim)
print("Shape:", sales.shape)

# Access Rows

print("\nFirst Row:", sales[0])
print("Second Row:", sales[1])

# Access Columns

print("First Column:", sales[:,0])
print("Second Column:", sales[:,1])

# Middle Element

print("Middle Element:", sales[1,1])

# Slice Rows

print("\nFirst Two Rows:")
print(sales[:2])

# Copy

sales_copy = sales.copy()

# View

sales_view = sales.view()

sales[0,0] = 999

print("\nModified Original:")
print(sales)

print("\nCopy:")
print(sales_copy)

print("\nView:")
print(sales_view)

# Reshape to 1D

print("\nFlattened:")
print(sales.reshape(-1))

# Join Arrays

extra = np.array([
    [1,1,1],
    [2,2,2],
    [3,3,3]
])

print("\nVertical Join")
print(np.vstack((sales, extra)))

print("\nHorizontal Join")
print(np.hstack((sales, extra)))

# Datatype

print("\nDatatype:", sales.dtype)
```
