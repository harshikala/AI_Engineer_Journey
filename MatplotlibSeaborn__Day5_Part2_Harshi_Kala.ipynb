### Exercise 8: HR Analytics

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(8)
n = 60
df8 = pd.DataFrame({
    'Employee': [f'Emp{i}' for i in range(1, n+1)],
    'Department': np.random.choice(['HR','IT','Finance','Sales'], n),
    'Performance': np.random.randint(1, 10, n),
    'Salary': np.random.randint(20000, 90000, n),
    'Experience': np.random.randint(0, 25, n),
    'Promotion': np.random.choice(['Yes','No'], n)
})
```

```python
new_salary = df8['Salary'].to_numpy() * 1.10
avg_performance = df8['Performance'].to_numpy().mean()
```

```python
promotion_stats = df8['Promotion'].value_counts()
dept_kpi = df8.groupby('Department')[['Performance','Salary']].mean()
top_performers = df8.sort_values('Performance', ascending=False).head(5)
```

```python
dept_kpi['Performance'].plot(kind='bar')
plt.title('Performance Chart')
plt.show()

plt.hist(df8['Salary'], bins=10, color='orange', edgecolor='black')
plt.title('Salary Distribution')
plt.show()
```

```python
sns.scatterplot(x='Performance', y='Salary', data=df8)
plt.title('Salary vs Performance')
plt.show()

sns.heatmap(df8[['Performance','Salary','Experience']].corr(), annot=True)
plt.show()
```

### Exercise 9: IPL Cricket Analysis

```python
np.random.seed(9)
n = 60
df9 = pd.DataFrame({
    'Player': [f'Player{i}' for i in range(1, n+1)],
    'Team': np.random.choice(['MI','CSK','RCB','KKR','SRH'], n),
    'Runs': np.random.randint(0, 800, n),
    'Wickets': np.random.randint(0, 30, n),
    'StrikeRate': np.round(np.random.uniform(100, 200, n), 2),
    'Matches': np.random.randint(1, 16, n)
})
```

```python
avg_runs = df9['Runs'].to_numpy().mean()
highest_sr = df9['StrikeRate'].to_numpy().max()
```

```python
team_runs = df9.groupby('Team')['Runs'].sum()
top_scorers = df9.sort_values('Runs', ascending=False).head(5)
bowling_stats = df9.groupby('Team')['Wickets'].sum()
```

```python
team_runs.plot(kind='bar')
plt.title('Runs by Team')
plt.show()

plt.bar(df9['Player'][:10], df9['StrikeRate'][:10])
plt.xticks(rotation=90)
plt.title('Strike Rate Comparison')
plt.show()
```

```python
sns.pairplot(df9[['Runs','Wickets','StrikeRate','Matches']])
plt.show()

sns.boxplot(x='Team', y='Runs', data=df9)
plt.show()
```

### Exercise 12: Manufacturing Quality Analysis

```python
np.random.seed(12)
n = 70
df12 = pd.DataFrame({
    'Machine': np.random.choice(['M1','M2','M3','M4'], n),
    'Production': np.random.randint(100, 1000, n),
    'Defects': np.random.randint(0, 50, n),
    'Shift': np.random.choice(['Morning','Evening','Night'], n),
    'Temperature': np.random.randint(20, 80, n),
    'Operator': np.random.choice(['Op1','Op2','Op3','Op4'], n)
})
```

```python
df12['Defect_Percentage'] = df12['Defects'] / df12['Production'] * 100
shift_production = df12.groupby('Shift')['Production'].sum()
operator_performance = df12.groupby('Operator')['Defect_Percentage'].mean()
machine_comparison = df12.groupby('Machine')['Production'].sum()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10,8))

shift_production.plot(kind='bar', ax=axs[0,0])
axs[0,0].set_title('Shift-wise Production')

machine_comparison.plot(kind='bar', ax=axs[0,1])
axs[0,1].set_title('Machine Comparison')

operator_performance.plot(kind='bar', ax=axs[1,0])
axs[1,0].set_title('Operator Defect %')

axs[1,1].hist(df12['Defect_Percentage'], bins=10, color='red', edgecolor='black')
axs[1,1].set_title('Defect % Distribution')

plt.tight_layout()
plt.show()
```

### Exercise 14: Smart City Traffic Analysis

```python
np.random.seed(14)
n = 80
df14 = pd.DataFrame({
    'Junction': np.random.choice(['J1','J2','J3','J4'], n),
    'Vehicles': np.random.randint(50, 1000, n),
    'Speed': np.random.randint(5, 80, n),
    'Congestion': np.random.choice(['Low','Medium','High'], n),
    'Time': np.random.randint(0, 24, n)
})
```

```python
peak_hours = df14.groupby('Time')['Vehicles'].sum().sort_values(ascending=False).head(5)
avg_speed = df14.groupby('Junction')['Speed'].mean()
congestion_count = df14['Congestion'].value_counts()
```

```python
peak_hours.plot(kind='bar')
plt.title('Peak Traffic Hours')
plt.show()

plt.hist(df14['Vehicles'], bins=10, color='green', edgecolor='black')
plt.title('Vehicle Distribution')
plt.show()

pivot14 = df14.pivot_table(values='Vehicles', index='Junction', columns='Congestion', aggfunc='sum')
sns.heatmap(pivot14, annot=True, fmt='.0f')
plt.show()
```

### Exercise 15: Financial Portfolio Analysis

```python
np.random.seed(15)
n = 40
df15 = pd.DataFrame({
    'Stock': [f'Stock{i}' for i in range(1, n+1)],
    'Investment': np.random.randint(10000, 100000, n),
    'CurrentValue': np.random.randint(8000, 150000, n),
    'Sector': np.random.choice(['IT','Banking','Pharma','Energy','FMCG'], n)
})
df15['Profit'] = df15['CurrentValue'] - df15['Investment']
```

```python
total_investment = df15['Investment'].to_numpy().sum()
total_profit = df15['Profit'].to_numpy().sum()
```

```python
sector_returns = df15.groupby('Sector')['Profit'].sum()
sector_returns.plot(kind='bar')
plt.title('Sector-wise Returns')
plt.show()

df15.groupby('Sector')['Investment'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title('Portfolio Distribution')
plt.show()

pivot15 = df15.pivot_table(values='Profit', index='Sector', aggfunc='sum')
sns.heatmap(pivot15, annot=True, fmt='.0f')
plt.title('Profit Heatmap')
plt.show()
```
