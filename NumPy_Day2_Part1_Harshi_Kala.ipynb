##Exercise 7: Copy and View
####Create a copy and modify the original array.
####Create a view and modify the original array.
####Use .base to identify ownership.

```python
arr = np.array([10, 20, 30, 40, 50])

print("Original Array:")
print(arr)
```

```python
# 1. Create a Copy and Modify the Original Array

copy_arr = arr.copy()

# Modify original array
arr[0] = 100

print("\nAfter Modifying Original Array:")
print("Original Array:", arr)
print("Copy Array:", copy_arr)

# Check ownership using .base
print("\nCopy Ownership:")
print(copy_arr.base)
```

```python
view_arr = arr.view()

# Modify original array
arr[1] = 200

print("\nAfter Modifying Original Array Again:")
print("Original Array:", arr)
print("View Array:", view_arr)

# Check ownership using .base
print("\nView Ownership:")
print(view_arr.base)
```

```python
print("\nMemory Ownership Check:")
print("Copy owns data:", copy_arr.base is None)
print("View owns data:", view_arr.base is None)
```

##Exercise 10: Joining Arrays
####Use concatenate(), stack(), hstack(), vstack() and dstack().
####Compare the output shapes.

```python
# Create sample arrays
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

print("Array 1:")
print(arr1)

print("\nArray 2:")
print(arr2)
```

```python
# 1. concatenate()
concatenated = np.concatenate((arr1, arr2))

print("\nUsing concatenate():")
print(concatenated)
print("Shape:", concatenated.shape)
```

```python
# 2. stack()
stacked = np.stack((arr1, arr2))

print("\nUsing stack():")
print(stacked)
print("Shape:", stacked.shape)
```

```python
# 2. stack()
stacked = np.stack((arr1, arr2))

print("\nUsing stack():")
print(stacked)
print("Shape:", stacked.shape)
```

```python
# 3. hstack()
hstacked = np.hstack((arr1, arr2))

print("\nUsing hstack():")
print(hstacked)
print("Shape:", hstacked.shape)
```

```python
# 4. vstack()
vstacked = np.vstack((arr1, arr2))

print("\nUsing vstack():")
print(vstacked)
print("Shape:", vstacked.shape)
```

```python
# 5. dstack()
dstacked = np.dstack((arr1, arr2))

print("\nUsing dstack():")
print(dstacked)
print("Shape:", dstacked.shape)
```

```python
# Comparing Output Shapes
print("\nComparison of Shapes:")
print("concatenate shape:", concatenated.shape)
print("stack shape:", stacked.shape)
print("hstack shape:", hstacked.shape)
print("vstack shape:", vstacked.shape)
print("dstack shape:", dstacked.shape)
```

##Exercise 11: Employee Dataset
####Perform indexing, slicing, reshape and joining operations on an employee dataset

```python
employees = np.array([
    [101, 25, 50000],
    [102, 30, 60000],
    [103, 28, 55000],
    [104, 35, 75000],
    [105, 40, 90000],
    [106, 32, 65000]
])

print("Employee Dataset:")
print(employees)

print("Shape:", employees.shape)
```

```python
print("\nFirst Employee Record:")
print(employees[0])

print("\nSalary of Third Employee:")
print(employees[2, 2])

print("\nEmployee ID of Fifth Employee:")
print(employees[4, 0])
```

```python
print("\nFirst Three Employees:")
print(employees[:3])

print("\nLast Two Employees:")
print(employees[-2:])

print("\nEmployee IDs:")
print(employees[:, 0])

print("\nEmployee Salaries:")
print(employees[:, 2])
```

```python
# Reshape dataset into 3D array
employee_3d = employees.reshape(2, 3, 3)

print("\nReshaped Employee Dataset (2,3,3):")
print(employee_3d)

print("Shape:", employee_3d.shape)
```

```python
# Create another employee dataset
new_employees = np.array([
    [107, 27, 52000],
    [108, 29, 58000]
])

print("\nNew Employee Dataset:")
print(new_employees)


# Concatenate rows
combined_employees = np.concatenate((employees, new_employees))

print("\nAfter concatenate():")
print(combined_employees)
print("Shape:", combined_employees.shape)


# Stack arrays vertically
vstack_employees = np.vstack((employees, new_employees))

print("\nAfter vstack():")
print(vstack_employees)
print("Shape:", vstack_employees.shape)


# Stack arrays horizontally
extra_data = np.array([
    [1],
    [2],
    [3],
    [4],
    [5],
    [6]
])

hstack_employees = np.hstack((employees, extra_data))

print("\nAfter hstack() (Adding Department ID):")
print(hstack_employees)
print("Shape:", hstack_employees.shape)
```

```python
# 2. Slicing Operations

# First two students
print("\nFirst Two Students Marks:")
print(marks[:2])

# Last two students
print("\nLast Two Students Marks:")
print(marks[-2:])

# First subject marks of all students
print("\nFirst Subject Marks:")
print(marks[:, 0])

# Second subject marks of all students
print("\nSecond Subject Marks:")
print(marks[:, 1])

# Third subject marks of all students
print("\nThird Subject Marks:")
print(marks[:, 2])
```

```python
# 3. Shape Operations

print("\nOriginal Shape:")
print(marks.shape)


# 4. Reshape Operations

# Reshape into 3D array
marks_3d = marks.reshape(2, 2, 3)

print("\nReshaped Marks Dataset (2,2,3):")
print(marks_3d)

print("Shape:", marks_3d.shape)
```

```python
# Reshape into 1D array
marks_1d = marks.reshape(-1)

print("\nFlattened Marks Array:")
print(marks_1d)

print("Shape:", marks_1d.shape)
```

```python
# 5. Joining Operations

# Create additional student marks
new_marks = np.array([
    [88, 91, 90],
    [76, 80, 85]
])

print("\nNew Student Marks:")
print(new_marks)
```

```python
# concatenate()
combined_marks = np.concatenate((marks, new_marks))

print("\nAfter concatenate():")
print(combined_marks)

print("Shape:", combined_marks.shape)
```
